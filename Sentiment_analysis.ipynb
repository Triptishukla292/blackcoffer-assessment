```python
def load_text_files(positive_file, negative_file):

  positive_sentences = []
  negative_sentences = []

  with open(positive_file, 'rb') as f:
    for line in f:
      encoding = chardet.detect(line)['encoding']
      positive_sentences.append(line.decode(encoding).strip())

  with open(negative_file, 'rb') as f:
    for line in f:
      encoding = chardet.detect(line)['encoding']
      negative_sentences.append(line.decode(encoding).strip())

  return positive_sentences, negative_sentences

# Load the positive and negative text files
positive_sentences, negative_sentences = load_text_files("/content/drive/MyDrive/New folder/master dictionary/positive-words.txt", "/content/drive/MyDrive/New folder/master dictionary/negative-words.txt")

# Print the positive and negative sentences
print("Positive sentences:")
print(positive_sentences[:5])
print("\nNegative sentences:")
print(negative_sentences[:5])
```

```python
# Define a function to calculate the positive score
def calculate_positive_score(text, positive_sentences):
    score = 0
    for word in text.split():
        if word in positive_sentences:
            score += 1
    return score

# Iterate over each row in the DataFrame
for index, row in final_df.iterrows():
    # Calculate the positive score for article titles
    article_title_positive_score = calculate_positive_score(row["article_title"], positive_sentences)

    # Calculate the positive score for article text
    article_text_positive_score = calculate_positive_score(row["article_text"], positive_sentences)

    # Combine the positive scores from article titles and article text
    #final_df.at[index, "article_title_positive_score"] = article_title_positive_score
    #final_df.at[index, "article_text_positive_score"] = article_text_positive_score
    final_df.at[index, "POSITIVE SCORE"] = article_title_positive_score + article_text_positive_score

# Print the DataFrame head
final_df.head()
```

```python
# Define a function to calculate the positive score
def calculate_negative_score(text, negative_sentences):

  score = 0
  for word in text.split():
    if word in negative_sentences:
      score -= 1
  return score * -1

# Iterate over each row in the DataFrame
for index, row in final_df.iterrows():
    # Calculate the negative_score for article titles
    article_title_negative_score = calculate_negative_score(row["article_title"], negative_sentences)

    # Calculate the negative_score for article text
    article_text_negative_score = calculate_negative_score(row["article_text"], negative_sentences)

    # Combine the positive scores from article titles and article text
    #final_df.at[index, "article_title_negative_score"] = article_title_negative_score
    #final_df.at[index, "article_text_negative_score"] = article_text_negative_score
    final_df.at[index, "NEGATIVE SCORE"] = article_title_negative_score + article_text_negative_score

# Print the DataFrame head
final_df.head()
```

```python
def calculate_polarity_score(positive_score, negative_score):
  denominator = positive_score + negative_score + 0.000001
  if denominator == 0:
    return 0
  else:
    return (positive_score - negative_score) / denominator

# Calculate the polarity score for each row in the DataFrame
final_df["POLARITY SCORE"] = final_df.apply(lambda row: calculate_polarity_score(row["POSITIVE SCORE"], row["NEGATIVE SCORE"]), axis=1)

# Print the DataFrame head
final_df.head()
```

```python
def count_syllables(word):
  count = 0
  vowels = "aeiouy"
  word = word.lower()
  if word[0] in vowels:
    count += 1
  for i in range(1, len(word)):
    if word[i] in vowels and word[i - 1] not in vowels:
      count += 1
  if word.endswith("e"):
    count -= 1
  if count == 0:
    count += 1
  return count

def count_complex_words(sentence):
  words = word_tokenize(sentence)
  complex_words = [word for word in words if count_syllables(word) > 2]
  return len(complex_words)
final_df["Number of Complex Words"] = final_df["combined_text"].apply(lambda sentence: count_complex_words(sentence))
```

```python

# Define a function to clean the text and count the words
def count_cleaned_words(text):
  # Tokenize the text
  words = word_tokenize(text)

  # Remove stop words and punctuations
  cleaned_words = [word for word in words if word.lower() not in stop_words_files and word not in string.punctuation]

  # Return the number of cleaned words
  return len(cleaned_words)

# Apply the function to the "Text" column and create a new column
final_df["Total Cleaned Words"] = final_df["combined_text"].apply(count_cleaned_words)

# Print the head of the DataFra
final_df["Total Cleaned Words"]
```

```python

def count_syllables(word):
  # Remove trailing "es" and "ed"
  word = word.lower().rstrip("es").rstrip("ed")

  # Count the number of vowels
  count = sum(1 for c in word if c in "aeiouy")

  # Handle exceptions
  if word.endswith("e"):
    count -= 1
  if count == 0:
    count = 1

  return count

# Create a new column to store the syllable count for each word
final_df["Syllable Count"] = 0

# Iterate through each row of the DataFrame
for index, row in final_df.iterrows():
  # Extract the text from the relevant column
  text = row["article_text"]

  # Split the text into words
  words = re.findall(r"\w+", text)

  # Count the syllables in each word and store it in the new column
  syllable_counts = [count_syllables(word) for word in words]
  final_df.loc[index, "Syllable Count"] = sum(syllable_counts)

# Print the DataFrame with the new column
final_df["Syllable Count"].head()
```

```python
# Define the function to count personal pronouns
def count_personal_pronouns(text):

  # Define the regular expression to match personal pronouns
  pronoun_regex = r"(?<!U\.S\.)\b(I|we|my|ours|us)\b"

  # Find all matches of the regular expression in the text
  matches = re.findall(pronoun_regex, text, flags=re.IGNORECASE)

  # Return the number of matches
  return len(matches)

# Create a new column to store the personal pronoun count for each text
final_df["Personal Pronoun Count"] = 0

# Iterate through each row of the DataFrame
for index, row in final_df.iterrows():
  # Extract the text from the relevant column
  text = row["article_text"]

  # Count the personal pronouns in the text and store it in the new column
  personal_pronoun_count = count_personal_pronouns(text)
  final_df.loc[index, "Personal Pronoun Count"] = personal_pronoun_count

# Print the DataFrame with the new column
final_df["Personal Pronoun Count"].head()
```

```python
def average_word_length(text):

  # Split the text into words
  words = text.split()

  # Calculate the total number of characters in all words
  total_characters = sum(len(word) for word in words)

  # Calculate the average word length
  average_length = total_characters / len(words)

  return average_length

# Create a new column to store the average word length for each text
final_df["Average Word Length"] = 0

# Iterate through each row of the DataFrame
for index, row in final_df.iterrows():
  # Extract the text from the relevant column
  text = row["article_text"]

  # Calculate the average word length for the text and store it in the new column
  final_df.loc[index, "Average Word Length"] = average_word_length(text)

# Print the DataFrame with the new column
final_df["Average Word Length"].head()
```
